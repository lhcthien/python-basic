```python
# import library
from pygame import mixer
from tkinter import *
# from tkinter import messagebox
from tkinter import messagebox as mb
from tkinter import ttk
import tkinter.font as font
import json
```

```python
#class to define the components of the GUI
class Quiz(Toplevel):
	# This is the first method which is called when a
	# new object of the class is initialized. This method
	# sets the question count to 0. and initialize all the
	# other methoods to display the content and make all the
	# functionalities available
	
	def __init__(self, master = None):
		super().__init__(master = master)
		self.geometry('800x450')        
		# set question number to 0
		self.q_no=0
		
		
		
		
		# assigns ques to the display_question function to update later.
		self.display_title()
		self.display_question()
		
		# opt_selected holds an integer value which is used for
		# selected option in a question.
		self.opt_selected=IntVar()
		
		# displaying radio button for the current question and used to
		# display options for the current question
		self.opts=self.radio_buttons()
		
		# display options for the current question
		self.display_options()
		
		# displays the button for next and exit.
		self.buttons()
		
		# no of questions
		self.data_size=len(question)
		
		# keep a counter of correct answers
		self.correct=0
		#self.label = Label(self)
		#self.label.place(x=35, y=15)       
		mixer.init()
		self.play_song()
# 		self.store_result()


	# This method is used to display the result
	# It counts the number of correct and wrong answers
	# and then display them at the end as a message Box
	def display_result(self):
		
		# calculates the wrong count
		wrong_count = self.data_size - self.correct
		correct = f"Correct: {self.correct}"
		wrong = f"Wrong: {wrong_count}"
		
		# calcultaes the percentage of correct answers
		score = int(self.correct / self.data_size * 100)
		result = f"Score: {score}%"
		
		# Shows a message box to display the result
		mb.showinfo("Result", f"{result}\n{correct}\n{wrong}")
		
		
		filename = 'data.json'
		
		with open(filename, "r") as f:  # reading a file
			data = json.load(f)  # deserialization
		data["correct"].append( f"Correct: {self.correct}")
		data["wrong"].append(f"Wrong: {wrong_count}")
		data["result"].append(f"Score: {score}%")
		
		
		with open(filename, 'w') as f:  #open the file in write mode
			json.dump(data, f)
			



	# This method checks the Answer after we click on Next.
	def check_ans(self, q_no):
		
		# checks for if the selected option is correct
		if self.opt_selected.get() == answer[q_no]:
			# if the option is correct it return true
			return True

	# This method is used to check the answer of the
	# current question by calling the check_ans and question no.
	# if the question is correct it increases the count by 1
	# and then increase the question number by 1. If it is last
	# question then it calls display result to show the message box.
	# otherwise shows next question.
	def next_btn(self):
		
		# Check if the answer is correct
		if self.check_ans(self.q_no):
			
			# if the answer is correct it increments the correct by 1
			self.correct += 1
		
		# Moves to next Question by incrementing the q_no counter
		self.q_no += 1
		
		# checks if the q_no size is equal to the data size
		if self.q_no==self.data_size:
			
			# if it is correct then it displays the score
			self.display_result()
# 			self.store_result()
			mixer.music.stop()
			# destroys the GUI
			self.destroy()
			self.quit()
		else:
			# shows the next question
			self.display_question()
			self.display_options()
			



	# This method shows the two buttons on the screen.
	# The first one is the next_button which moves to next question
	# It has properties like what text it shows the functionality,
	# size, color, and property of text displayed on button. Then it
	# mentions where to place the button on the screen. The second
	# button is the exit button which is used to close the GUI without
	# completing the quiz.
	def buttons(self):
		
		# The first button is the Next button to move to the
		# next Question
		next_button = Button(self, text="Next",command=self.next_btn,
		width=10,bg="blue",fg="white",font=("ariel",16,"bold"))
		
		# placing the button on the screen
		next_button.place(x=350,y=380)
		
		# This is the second button which is used to Quit the GUI
		quit_button = Button(self, text="Quit", command=self.quit,
		width=5,bg="black", fg="white",font=("ariel",16," bold"))
		
		# placing the Quit button on the screen
		quit_button.place(x=700,y=50)
		
	def quit(self):
		self.destroy()
		mixer.music.stop()
		question.clear()
		answer.clear()
		options.clear()
		song.clear()

	# This method deselect the radio button on the screen
	# Then it is used to display the options available for the current
	# question which we obtain through the question number and Updates
	# each of the options for the current question of the radio button.
	def display_options(self):
		val=0
		
		# deselecting the options
		self.opt_selected.set(0)
		
		# looping over the options to be displayed for the
		# text of the radio buttons.
		for option in options[self.q_no]:
			self.opts[val]['text']=option
			val+=1

	def play_song(self):
		mixer.music.load(song[self.q_no])
		
		mixer.music.play(-1)
			
			
			

	# This method shows the current Question on the screen
	def display_question(self):
		
		# setting the Question properties
		q_no = Label(self, text= question[self.q_no], width=60,
		font=( 'ariel' ,16, 'bold' ), anchor= 'w' )
		
		#placing the option on the screen
		q_no.place(x=70, y=100)
		self.play_song()


	# This method is used to Display Title
	def display_title(self):
		
		# The title to be shown
		title = Label(self, text="Music guessing game",
		width=50, bg="green",fg="white", font=("ariel", 20, "bold"))
		
		# place of the title
		title.place(x=0, y=2)


	# This method shows the radio buttons to select the Question
	# on the screen at the specified position. It also returns a
	# list of radio button which are later used to add the options to
	# them.
	def radio_buttons(self):
		
		# initialize the list with an empty list of options
		q_list = []
		
		# position of the first option
		y_pos = 150
		
		# adding the options to the list
		while len(q_list) < 4:
			
			# setting the radio button properties
			radio_btn = Radiobutton(self,text=" ",variable=self.opt_selected,
			value = len(q_list)+1,font = ("ariel",14))
			
			# adding the button to the list
			q_list.append(radio_btn)
			
			# placing the button
			radio_btn.place(x = 100, y = y_pos)
			
			# incrementing the y-axis position by 40
			y_pos += 40
		
		# return the radio buttons
		return q_list  
```

```python
roota = Tk()
roota.title('MUSIC GUESSING GAME')
roota.geometry("800x450")
bg = PhotoImage(file = 'Webp.net-resizeimage.png')
label = Label(roota, image=bg)
label.place(x=0, y=0)

# Title
title = Label(roota,text="What is this song?",bd=9,relief=GROOVE,
             font=("times new roman",50,"bold"),bg="white",fg="green")
title.pack()

def start():
    if bool(question):
        tk = Toplevel(roota)
        tk.geometry('300x200')
        tk.resizable(False, False)
        tk.title('Choose name')
        tk.grab_set()
        label = Label(tk, text="Please select your player name:",
                  font=("times new roman",15,"bold"))
        label.pack(padx=5, pady=5)
        player_name = Entry(tk)
        player_name.pack()

        def playername_enter(event):
            playername_select()    
        player_name.bind('<Return>', playername_enter)

        def playername_select():
            reply = mb.askyesno(
                title='Get ready',
                message=f'You selected {player_name.get()}!'
            )
            if reply==True:
                with open(f'data.json', 'r') as f:
                    data = json.load(f)
                data['player'].append(player_name.get())
                with open('data.json', 'w') as f:  #open the file in write mode
                    json.dump(data, f)
                tk.destroy()
                quiz = Quiz(roota)
    # Button
        ok_btn = Button(tk, text='''OK, Let's go''', command=playername_select
                    ,bg="#ef9007",fg="white",
                    font=("ariel",13,"bold"))
        ok_btn.pack()
    
        back_btn = Button(tk, text='Back', command=tk.destroy
                    ,bg="#1e3f54",fg="white",
                    font=("ariel",13,"bold"))
        back_btn.pack()
    else: mb.showwarning('A little note', 'Please choose a mixtape first!')
    

def mixtape():
    tk = Toplevel(roota)
    # config the tk window
    tk.geometry('300x200')
    tk.resizable(False, False)
    tk.title('Mixtape pick')
    tk.grab_set()
    # label
    label = Label(tk, text="Please select a mixtape:",font=("times new roman",15,"bold"))
    label.pack(padx=5, pady=5)

    # create a combobox
    selected_mixtape = StringVar()
    # selected_mixtape.set('Billboard US Hit')
    mixtape_cb = ttk.Combobox(tk, textvariable=selected_mixtape)
    
    # get first 3 letters of every month name
    mixtape_cb['values'] = ['Billboard US Hit', 'Indie UK Pop', 'Vpop', 'Kpop']

    # prevent typing a value
    mixtape_cb['state'] = 'readonly'

    # place the widget
    mixtape_cb.pack(padx=5, pady=5)
      
    # bind the selected value changes
    def mixtape_select(event):
        """ handle the mixtape selected event """
        global question
        global options
        global answer
        global song
        select = mb.askyesno(
            title='Result',
            message=f'You selected {selected_mixtape.get()}!'
        )
        if select == True:
            with open(f'{selected_mixtape.get()}.json') as f:
                data = json.load(f)
            question = (data['question'])
            options = (data['options'])
            answer = (data['answer'])
            song = (data['song'])
            # Store mixtape
            with open(f'data.json', 'r') as f:
                data = json.load(f)
            data['mixtape'].append(selected_mixtape.get())
            with open('data.json', 'w') as f:  #open the file in write mode
                json.dump(data, f)
    mixtape_cb.bind('<<ComboboxSelected>>', mixtape_select)
    
    ok_btn = Button(tk, text='OK', command=tk.destroy,bg="indigo",fg="white",
                    font=("ariel",13,"bold"))
    ok_btn.pack()
    back_btn = Button(tk, text='Back', command=tk.destroy
                    ,bg="#1e3f54",fg="white",
                    font=("ariel",13,"bold"))
    back_btn.pack()

    
def score():
    tk = Toplevel(roota)
    tk.geometry("800x450")
    title = Label(tk, text="Scoreboard",
        width=50, bg="#f6d317",fg="#1e3f54", font=("ariel", 20, "bold"))
    title.pack()
    back_button = Button(tk, text="Back", command=tk.destroy,
                        width=5,bg="black", fg="white",font=("ariel",16," bold"))

    back_button.place(x=0,y=70)
    with open("data.json", "r") as f:
        data = json.load(f)
    list_correct = data["correct"]
    list_wrong = data["wrong"]
    list_result = data["result"]
    list_player = data["player"]
    list_mixtape = data["mixtape"]
    player_lb = Label(tk, text='Player', font=("ariel", 15, 'bold'), fg="#1d1918")
    mixtape_lb = Label(tk, text='Mixtape', font=("ariel", 15, 'bold'), fg="#1d1918")
    player_lb.place(x=190, y=70)
    mixtape_lb.place(x=290, y=70)
    row = 100
    col = 100
    for i in range(len(list_result)):
        Label(tk, text=f"Round {i+1}:",font=("ariel", 15)).place(x=row, y=col)
        Label(tk, text=list_player[i],font=("ariel", 15)).place(x=row+90, y=col)
        Label(tk, text=list_mixtape[i],font=("ariel", 15)).place(x=row+170, y=col)
        Label(tk, text=list_correct[i],font=("ariel", 15)).place(x=row+330, y=col)     
        Label(tk, text=list_wrong[i],font=("ariel", 15)).place(x=row+435, y=col)
        Label(tk, text=list_result[i],font=("ariel", 15)).place(x=row+530, y=col)
        col += 30
        

greet = Label(roota, text = "Welcome to music guessing game, please choose a mixtape before playing!",
              font=("ariel", 15, "bold"))
greet.pack()

start_btn = Button(roota, text="Start",command=start,
        width=10,bg="#883d94",fg="white",font=("ariel",16,"bold"))
mixtape_btn = Button(roota, text="Mixtape",command=mixtape,
        width=10,bg="#ffffff",fg="black",font=("ariel",16,"bold"))
score_btn = Button(roota, text="Score",command=score,
        width=10,bg="#cccccc",fg="white",font=("ariel",16,"bold"))
quit_btn = Button(roota, text="Quit", command=roota.destroy,
        width=5,bg="black", fg="white",font=("ariel",16," bold"))

#start_btn.bind("<Button>", lambda e: Quiz(roota))

start_btn.place(x=333,y=150)
mixtape_btn.place(x=333,y=200)
score_btn.place(x=333,y=250)
quit_btn.place(x=365,y=300)


roota.mainloop()
```
