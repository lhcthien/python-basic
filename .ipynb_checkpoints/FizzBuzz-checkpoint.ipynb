# Fizz Buzz

```python
n = input('Nhap so bat dau va ket thuc: ')
x = n.split()
start = int(x[0])
end = int(x[1])

if start>end:
    print('Start number must be smaller than end number')
    n = input('Nhap so bat dau va ket thuc: ')
    x = n.split()
    start = int(x[0])
    end = int(x[1])
else: print('''Everything's ok''')

for i in range(start, end + 1):
    if i % 3 == 0 and i % 5 == 0:
        print("fizzbuzz")
    elif i % 5 == 0:
        print("buzz")
    elif i % 3 == 0:
        print("fizz")
    else:
        print(i)
```
